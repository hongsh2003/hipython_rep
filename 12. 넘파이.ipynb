## 단위행렬과 대각행렬의 생성

np.identity(): 단위행렬 생성\
np.diag(): 대각행렬 생성\
np.eye() 함수: 대각선이 1이고 나머지가 0인 2차원 배열을 생성하는 함수
- identity() 함수와 유사하지만, 직사각형 행렬을 만들 수 있고 대각선의 위치를 조절할 수 있다는 점에서 더 유연


```python
# 3x3 단위행렬 생성
identity_matrix = np.identity(3)
print(identity_matrix)
```

```python
# 대각원소가 [1, 2, 3]인 대각행렬 생성
diagonal_matrix = np.diag([1,2,3])
print(diagonal_matrix)
```

```python
a = np.arange(9).reshape (3,3)
diagonal_matrix_value = np.diag(a)
print(diagonal_matrix_value)
```

```python
# 3x3 단위행렬 생성 (N=3)
eye_matrix = np.eye(3)
eye_matrix
```

```python
# 4x3 직사각형 행렬 생성 (N=4, M=3)
rect_eye = np.eye(4,3)
rect_eye
```

```python
# k 매개변수로 대각선 위치 조정 (k=1은 주대각선 위로 한 칸)
shifted_eye = np.eye(3,k=1)
shifted_eye
```

## 인덱싱 연습하기

```python
import numpy as np

# 기준이 될 4x4 배열 생성 (이미지와 동일하게 0부터 15까지의 숫자로 채움)
arr_2d = np.arange(16).reshape((4, 4))

print("원본 4x4 배열:")
print(arr_2d)
print("-" * 30)
```

```python
# 첫 번째 행 전체
print(arr_2d[:1,])
# 강사님 답: arr_2d[0]
```

```python
#마지막 열 전체
print(arr_2d[:,3])
# 강사님 답: arr_2d[:,-1]
```

```python
#첫 번째 행과 마지막 행만 선택, 각 행의 첫번째, 마지막 열 선택
# 연속 인덱싱 => 겹쳐지는 부분 => 교차인덱싱
print(arr_2d[::3,::3])
# 강사님 답: arr_2d[[0,3]][:,[0,3]]
```

```python
# 짝수 인덱스 행(0번째, 2번째 행), 홀수 인덱스 열 선택
print(arr_2d[::2, 1::2])
# 강사님 답: arr_2d[::2][:,1::2]
```

```python
#위의 3차원 행렬을 작성하세요
a1=np.arange(36).reshape(6,6)

a2=np.stack(((a1,a1,a1,a1)))
a2
```

```python
# 2번째 층 선택
print(a2[1,:,:])
```

```python
# 각 층의 (0,0) 위치의 값을 각 층의 인덱스로 설정
for d_idx in range (4):
    a2[d_idx, 0,0] = d_idx
a2
```

```python
#모든 층, 네번째행, 두번째 열 원소 선택, 1차원 배열로 작성
print(a2[:,3,1])
```

```python
#배열의 첫 두 층, 마지막 두 행, 마지막 두 열 선택
print(a2[:2,4:,4:])
# 강사님 답 a2[:2, -2:, -2:]
```

```python
#짝수 인덱스층, 모든행의  세번째열 선택
print(a2[1::2,:,2])
# 강사님 답 a2 [::, :, 2]
```

```python
#세번째 층, 두번째 행에서 네번째 행, 홀수 인덱스 열
print(a2[2,1:4,1::2])
```
