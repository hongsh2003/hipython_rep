# 인덱싱 연습하기

```python
import numpy as np

# 기준이 될 4x4 배열 생성 (이미지와 동일하게 0부터 15까지의 숫자로 채움)
arr_2d = np.arange(16).reshape((4, 4))

print("원본 4x4 배열:")
print(arr_2d)
print("-" * 30)
```

```python
# 첫 번째 행 전체
arr_2d[0 , : ] 
```

```python
#마지막 열 전체
arr_2d[: , -1]
```

```python
arr_2d
```

```python
#첫 번째 행과 마지막 행만 선택, 각 행의 첫번째 열, 마지막 열 선택
#연속 인덱싱 > 겹쳐지는 부분 > 교차인덱싱

arr_2d[[0,3]][:, [0,3]]
```

```python
# 짝수 인덱스 행(0번째, 2번째 행), 홀수 인덱스 열(1,3) 선택
arr_2d[::2, 1::2]
```

```python
arr_2d[::2][:,1::2]
```

```python
#위의 3차원 행렬을 작성하세요
arr_3d = np.arange(36).reshape((6,6))
arr_3d
```

```python
np.vstack([arr_3d, arr_3d, arr_3d]).shape
```

```python
arr_3d_new = np.stack([arr_3d, arr_3d, arr_3d, arr_3d])
arr_3d_new.shape
```

```python
# 2번째 층(깊이, 면) 선택
arr_3d_new[1, :, :]
```

```python
#각 층의 (0,0)위치의 값을 각 층의 인덱스로 설정

for d_idx in range(4):
  arr_3d_new[d_idx, 0,0] = d_idx
arr_3d_new
```

```python
#모든 층, 네번째 행, 두번째 열 원소 선택, 1차원 배열로 작성
#arr_3d_new[:, 0, 0]

arr_3d_new[:, 3, 1]
```

```python
#배열의 첫 두 층, 마지막 두 행, 마지막 두 열 선택
```

```python
arr_3d_new[:2, -2:, -2:]
```

```python
#짝수 인덱스층, 모든행의 세번째열 선택
arr_3d_new[::2, :, 2]
```

```python
#세번째 층, 두번째 행에서 네번째 행, 홀수 인덱스 열
arr_3d_new[2, 1:4, 1::2]
```
