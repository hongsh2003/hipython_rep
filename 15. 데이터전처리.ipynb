# melt( )

- id_vars: 유지할 변수 리스트 (튜플, 행렬)
- value_vars: 피봇 되돌리기 값으로 쓰일 열 목록
- var_name: 만들어지는 새 컴럼의 이름
- value_name: 새 컬럼의 값의 컬럼 이름

```python
pew_df.melt(id_vars = 'religion')
```

```python
pew_df.melt(id_vars ='religion',
            var_name='income',
            value_name='count'
            )
```

```python
ebola_df = pd.read_csv('./data/country_timeseries.csv')
ebola_df.info()
```

```python
ebola_df.iloc[:5,[0,1,2,10]]
```

```python
# 1. 값 Cases~, Death~ id: Date, Day
ebola_long = ebola_df.melt(id_vars=['Date','Day'])
ebola_long
```

```python
#2. column 값을 2개로 분할
variable_split = ebola_long['variable'].str.split('_')
type(variable_split)
```

```python
variable_split[:5]
```

```python
variable_split[0][0]
```

```python
status_values = variable_split.str.get(0) #cases
country_values = variable_split.str.get(1) # country
type(status_values)
```

```python
status_values
```

```python
country_values
```

```python
#3. 컬럼추가
ebola_long['status']=status_values
ebola_long['country']=country_values
ebola_long
```

```python
weather = pd.read_csv('./data/weather.csv')
weather
```

```python
weather.columns
```

```python
#1) day로 언피봇 melt()
weather_melt = weather.melt(
            id_vars = ['id','year','month','element'],
            var_name = 'day',
            value_name = 'temp'
)

print(weather_melt)
```

```python
#2) element를 피봇팅 > tmax, tmin 컬럼 생성
weather_tidy = weather_melt.pivot_table(
                index = ['id', 'year', 'month', 'day'],
                columns = 'element',
                values = 'temp'
                ).reset_index()

weather_tidy.head()
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat
```
