```python
import requests
import folium
from pandas import json_normalize
```

```python
targetSite='https://www.starbucks.co.kr/store/getStore.do?r=8K3XF06R9Q'
req=requests.post(targetSite, data={
'ins_lat': 37.563398,
'ins_lng': 126.9863309,
'p_sido_cd': '01',
'p_gugun_cd': '',
'in_biz_cd': '',
'iend': 600,
'set_date': ''
})
starbucks=req.json()
print(type(starbucks))
```

```python
st_df = json_normalize(starbucks, 'list')
st_df
```

```python
st_df_map = st_df[['s_name','sido_code','sido_name','gugun_code','gugun_name',
'doro_address','lat','lot']]

st_df_map
```

```python
st_df_map.info()
```

```python
st_df_map['lat'] = st_df_map['lat'].astype(float)
st_df_map['lot'] = st_df_map['lot'].astype(float)

st_df_map.info()
```

```python
st_df_map.head()
```

```python
loc = [37.501087, 127.043069]
stMap = folium.Map(location=loc, zoom_start=18)
pop = folium.Popup('역삼아레나빌딩', max_width=100)
folium.Marker(location=loc, popup=pop).add_to(stMap)
stMap
```

```python
# folium 지도 객체를 한 번만 생성
stmap = folium.Map(location=[st_df_map['lat'].mean(), st_df_map['lot'].mean()], zoom_start=14)

# DataFrame의 모든 좌표 반복해서 마커 추가
for idx, row in st_df_map.iterrows():
    folium.Marker(
        location=[row['lat'], row['lot']], # 위도(lat), 경도(lon)
        popup=folium.Popup('스타벅스', max_width=100)
    ).add_to(stmap)

stmap
```
